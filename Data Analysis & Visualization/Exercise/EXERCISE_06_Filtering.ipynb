# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Find the books written by Pete Souza
books=pd.read_csv('../data/bestsellers.csv')

pete=books['Author']=='Pete Souza'

books[pete]
```

```python
# Find the books that are under 10 dollars

under_10=books['Price']<10

books[under_10]
```

```python
#Find the books that have a price between 50 and 60 dollars

book_between=(books['Price'] >=50) & (books['Price'] <=60)

books[book_between]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens

books[books.Author.isin(['Kristin Hannah','Andy Weir','Delia Owens'])]
```

```python
#Find the Non Fiction books that are rated 4.9

non_fiction=books["Genre"]=='Non Fiction'
user_rating=books['User Rating']>=4.9

books[non_fiction & user_rating]
```

```python
#Find the fiction book with the lowest User Rating
books[~non_fiction].sort_values('User Rating').head()
```

```python
#Find 2012's top 5 Fiction books with the most Reviews

book_2012=books['Year']==2012

books[book_2012 & ~non_fiction ].sort_values(by="Reviews",ascending=False).head()
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5

# count the number of books for each author with a rating under 4.5
low_rated_books = books[books['User Rating'] < 4.5]['Author'].value_counts().head()

# create a bar plot
low_rated_books.plot(kind='bar')
plt.title('Authors with the most books with a rating under 4.5')
plt.xlabel('Author')
plt.ylabel('Number of books')
plt.show()
```
