## Pandas Filtering Data

```python
import pandas as pd
houses = pd.read_csv("data/kc_house_data.csv")
titanic = pd.read_csv("data/titanic.csv")
netflix = pd.read_csv("data/netflix_titles.csv", sep="|", index_col=0)
```

```python
df = titanic.head()
df.sex 
```

```python
# conver age to int

titanic.age
```

```python
df[df.sex == 'female']
```

```python
titanic[titanic.age == "18"]
```

## People from New York, NY

```python
titanic[titanic['home.dest']=='New York, NY']
```

### only people survived whose age is 18

```python
survived=titanic[titanic.survived == 1 ] # survived

survived[titanic['age']=='18']
```

```python
## house price > 5M
houses[houses.price > 5000000]
```

```python
# bedrooms > 10

houses[houses.bedrooms >= 10].sort_values(by='bedrooms', ascending=False)
```

## Between Method

```python
# homes between 5 to 7 bedrooms

houses[houses.bedrooms.between(5,7)]
```

## Isin() Method

```python

countries=['India','Japan',"China"]
# only show movies from these countries

netflix[netflix.country.isin(countries)]
```

```python
matured_shows=netflix['rating'].isin(['TV-MA', 'TV-14'])

netflix[matured_shows]
```

## Combining conditional statements (and & or)

```python
# AND (&) operator
houses[(houses['waterfront']==1) & (houses['view']>=4)]
```

```python
# OR (|) operator
houses[(houses['waterfront']==1) | (houses['view']>=4)]
```

```python
# matured_shows release year in 2019

netflix[matured_shows & (netflix['release_year']==2012)]

```
