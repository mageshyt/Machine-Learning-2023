## Pandas Filtering Data

```python
import pandas as pd
houses = pd.read_csv("data/kc_house_data.csv")
titanic = pd.read_csv("data/titanic.csv")
netflix = pd.read_csv("data/netflix_titles.csv", sep="|", index_col=0)
```

### only people survived whose age is 18

```python
## house price > 5M
houses[houses.price > 5000000]
```

```python
# bedrooms > 10

houses[houses.bedrooms >= 10].sort_values(by='bedrooms', ascending=False)
```

## Between Method

```python
# homes between 5 to 7 bedrooms

houses[houses.bedrooms.between(5,7)]
```

## Isin() Method

```python

countries=['India','Japan',"China"]
# only show movies from these countries

netflix[netflix.country.isin(countries)]
```

```python
matured_shows=netflix['rating'].isin(['TV-MA', 'TV-14'])

netflix[matured_shows]
```

## Combining conditional statements (and & or)

```python
# AND (&) operator
houses[(houses['waterfront']==1) & (houses['view']>=4)]
```

```python
# OR (|) operator
houses[(houses['waterfront']==1) | (houses['view']>=4)]
```

```python
# matured_shows release year in 2019

netflix[matured_shows & (netflix['release_year']==2012)]

```

```python
# high quality houses
high_quality=(houses['grade'] >= 11) # 😙

# good view
good_view=houses['view']==4

# under 3000 sq

under_3000=houses['sqft_living'] <= 3000
houses[high_quality & good_view & under_3000]
```

```python
# houses built after 2014
house_build_after14=houses['yr_built'] >=2014

# houses renovated after 2014

house_renovated_after14=houses['yr_renovated'] >=2014


# select houses built after 2014 and renovated after 2014

recent_house=houses[house_build_after14 | house_renovated_after14]

recent_house.sort_values(by='yr_renovated')


```

```python
# movies directed by david fincher or martin scorsese
fincher=netflix['director']=='David Fincher'
scorsese=netflix['director']=='Martin Scorsese'
# method 1
netflix[fincher | scorsese]

# method 2

netflix[netflix['director'].isin(['David Fincher', 'Martin Scorsese'])]
```

# bitwise operators

```python
df=titanic.head()
women=df.sex == 'female'
# negation ~
df[~women] # this will list which are not women gender


# people who not survived

not_survived=titanic.survived == 0

titanic[not_survived]

# find survied using negation

titanic[~not_survived]
```

 check for missing values using isna()

 check for non-missing values using notna()


```python
# find the movies which has no director

netflix[netflix['director'].isna()]
```

```python
# 🔎 find the movies which has no director and cast

no_director=netflix['director'].isna()
no_cast=netflix['cast'].isna()

netflix[no_director & no_cast]

# find the movies which has director or cast


netflix[netflix['director'].notna() | netflix['cast'].notna()]
```
