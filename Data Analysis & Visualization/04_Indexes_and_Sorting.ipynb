## Sorting

## Sort by multiple columns

```python
house_prices = pd.read_csv("data/kc_house_data.csv")
titanic = pd.read_csv("data/titanic.csv")
```

```python
house_prices.sort_values(by="price",ascending=False)
```

```python
sorted_country=world_happiness.sort_values(by=["Country name"])
sorted_country.head()
```

#### here we are sorting by the 'price' in descending order and then by the 'bedrooms' in ascending order

```python
house_prices.sort_values(by=["price","bedrooms"],ascending=[False,True])
```

# sort the text column in ascending order (case-sensitive)

```python
titanic.sort_values("name",key=lambda x:x.str.lower())
```

```python
titanic.sort_values(["name","age"],ascending=[True,False])
```

## Sorting and Plotting 🧑‍🚀

```python
titanic.pclass.value_counts().sort_values()
```

```python
titanic["sex"].value_counts().sort_index(ascending=False).plot(kind="bar")
```

```python
house_prices["bedrooms"].value_counts().sort_index().plot(kind="bar")
```

## loc[] & iloc[]

```python
# loc - label based indexing

world_happiness.loc["India"]
```

```python
world_happiness.loc[['India','Pakistan',"Japan"]]
```

## Range of loc value

```python
titanic.loc[0:10] 
```

### lets learn about Iloc

1. **iloc** -> integer location based indexing
2. **loc** -> label based indexing

```python
sorted_country.iloc[0:10]
```

### [iloc & loc] with custom column

```python
house_prices.loc[2000:2005,['price','bedrooms']]
```

```python
sorted_country.loc['India':'Japan',['Regional indicator','Ladder score']]
```

```python
titanic.iloc[0:10,0:3]
```

```python
titanic["age"].value_counts().iloc[0:10].plot(kind="bar")
```
