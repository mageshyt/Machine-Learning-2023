# Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
import  numpy as np
import pandas as pd
```

```python
# direct plot [state less plot]
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, 'o', color='blue', )
x[-1]
```

```python
# method 1
fig = plt.figure() # an empty figure with no axes
ax = fig.add_subplot() # a set of axes
ax.plot(x, y, 'o', color='red' )

```

```python
# method 2

fig=plt.figure() # an empty figure with no axes

ax1=fig.add_axes([1,1,1,1]) # 2x2 grid, first subplot


ax1.plot(x, y, 'o', color='red' )

plt.show()
```

```python
# method 3 (recommended)

fig, ax = plt.subplots() # a figure with a single axes

ax.set(
    title='Simple plot',
    xlabel='x-axis',
    ylabel='y-axis'

)
ax.plot(x, y, 'o', color='red' )
```

# Matplotlib example workflow

```python
# 1. prepare data
x = np.array([1,2,3,4])
y = np.array([11,22,33,44])

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,5)) # width, height

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(
    title='Simple plot',
    xlabel='x-axis',
    ylabel='y-axis'
)

# 5. save & show (you save the whole figure)
plt.show()
```

# ploting from pandas dataframe

```python

mavel_df=pd.read_csv("./csv/Marvel_DC_imdb.csv")
```

```python
# lets clean the data

mavel_df.info()

print("size of the data",mavel_df.size)
# drop the null value
mavel_df.dropna(inplace=True)

print("size of the data",mavel_df.size)

mavel_df

```

```python
# get  revenue 
movie_revenue=mavel_df["USA_Gross"]

price_conver={
    "M":1000000,
    "K":1000
}
# remove $ and ,
mavel_df["USA_Gross"]=mavel_df["USA_Gross"].apply(lambda x:x.replace(",","").replace("$","") )
# covert M to 1000000 and K to 1000
mavel_df
```

```python
# sort the data and get top 10
mavel_df.sort_values(by=["USA_Gross"],ascending=False,inplace=True)

# plot the data

fig,ax=plt.subplots( nrows=2,ncols=2 ,figsize=(40,10))

ax[0,0].scatter(mavel_df["Movie"][:10],mavel_df["USA_Gross"][:10]) # scatter plot 

ax[0,0].set(title="Top 10 movie by revenue",xlabel="Movie name",ylabel="Revenue in $")

ax[0,1].bar(mavel_df["Movie"][:10],mavel_df["USA_Gross"][:10]) # bar plot

ax[0,1].set(title="Top 10 movie by revenue",xlabel="Movie name",ylabel="Revenue in $")

ax[1,0].plot(mavel_df["Movie"][:10],mavel_df["USA_Gross"][:10]) # line plot
ax[1,0].set(title="Top 10 movie by revenue",xlabel="Movie name",ylabel="Revenue in $")

ax[1,1].hist(mavel_df["USA_Gross"][:10]) # histogram
ax[1,1].set(title="Top 10 movie by revenue",xlabel="Movie name",ylabel="Revenue in $")
```

```python
# total revenue of marvel

mavel_df["total_revenue"]=mavel_df["USA_Gross"].cumsum()

```

```python
fig,ax=plt.subplots(figsize=(20,10))

# only 20 movie
ax.bar(mavel_df["Year"][:10],mavel_df["total_revenue"][:10])

ax.set(title="Total revenue of marvel",xlabel="Year",ylabel="Revenue in $")
# plot the data
plt.show()
```


## which one should we use ? (Pyplot or matplotlib OO method ?)

* When plotting something quickly ,then we can use pyplot method


* when plotting something advance ,use the OO method oo --> objects oriented method

```python
# movie revenue over 100M
over_400M=mavel_df[mavel_df["USA_Gross"].astype(int)>400000000]
```

```python
# pyplot method
over_400M.plot(kind='scatter',x="Year",y="USA_Gross",figsize=(5,5),);
```

```python
# OO method
fig,ax=plt.subplots(figsize=(5,5))

over_400M.plot(kind='scatter',x="Year",y="USA_Gross",ax=ax,colorbar=True,marker="o",c="USA_Gross",figsize=(10,10) );

ax.set(title="Movie revenue over 400M",xlabel="Movie",ylabel="Revenue in $")

plt.show()
```

### another Example


```python
# heart disease data

heart_disease=pd.read_csv("./csv/heart-disease.csv")
```

```python

# plot the data

fig,ax=plt.subplots(figsize=(10,5))

scatter=ax.scatter(x=heart_disease["age"],y=heart_disease["chol"],c=heart_disease["target"],cmap="winter")

ax.set(title="Heart disease",xlabel="Age",ylabel="Cholestrol")

ax.legend(*scatter.legend_elements(),title="Target")

ax.axhline(heart_disease["chol"].mean(),linestyle="--");
```

```python
# sbuplot of chol,age,thalach

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

# add data to ax0

ax0.scatter(x=heart_disease["age"],y=heart_disease["chol"],c=heart_disease["target"],cmap="winter")

# add details to ax0

ax0.set(title="Heart disease",xlabel="Age",ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target")

# mean line

ax0.axhline(heart_disease["chol"].mean(),linestyle="--",color="red",);
# length for mean line

ax0.axhspan(heart_disease["chol"].mean(),heart_disease["chol"].max(),alpha=0.2,color="green")

# add data to ax1

ax1.scatter(x=heart_disease["age"],y=heart_disease["thalach"],c=heart_disease["target"],cmap="winter")

# add details to ax1

ax1.set(title="Heart disease",xlabel="Age",ylabel="thalach")

ax1.legend(*scatter.legend_elements(),title="Target")

# mean line

ax1.axhline(heart_disease["thalach"].mean(),linestyle="--",color="red");

ax1.axhspan(heart_disease["thalach"].mean(),heart_disease["thalach"].max(),alpha=0.2,color="blue")

# add title to the figure
fig.suptitle("Heart disease analysis",fontsize=16,fontweight="bold");

```
